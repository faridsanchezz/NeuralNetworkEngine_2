# Dataset MNIST
789 features
10 classes

```python
path_mnist_data = "C:/Users/OSMEL/Documents/OH/data/data"
X_train, X_test, y_train, y_test = load_and_preprocess_data_mnist(path_mnist_data)

print("Shape of X:", X_train.shape)
print("Shape of y:", y_train.shape)
print("Shape of y:", X_test.shape)
print("Shape of y:", y_test.shape)
```

### Prueba con optimizador ADAM y MSE

```python
# to initialize the weights you can use different options: he, xavier, normal
layers = [
    DenseLayer(input_size=784, output_size=128, init_method='he'),
    ReLU(),
    DenseLayer(input_size=128, output_size=64, init_method='he'),
    ReLU(),
    DenseLayer(input_size=64, output_size=10, init_method='he'),
    Softmax()
]

network = NeuralNetwork(
        layers=layers,
        loss_function=MSE(),
        optimizer=Adam(learning_rate=0.001)
    )


training_history = train_network(
    network=network,
    X=X_train,
    y=y_train,
    epochs=100,
    batch_size=100,
    early_stopping_patience=5,
    validation_split=0.2,
    verbose=2
)


y_pred = network.predict(X_test)
print(evaluate_model(y_pred, y_test))
confusionMatrix(y_test, y_pred)
display_training_history(training_history)
```
